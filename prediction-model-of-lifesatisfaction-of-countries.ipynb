```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
```

# Reading the DATA

```python
data_root="https://raw.githubusercontent.com/ageron/data/main/lifesat/lifesat.csv"
life_sat=pd.read_csv(data_root)
```

```python
life_sat.shape
```

```python
life_sat.head()
```

```python
GDP=life_sat[["GDP per capita (USD)"]].values
```

```python
life_sat_factor=life_sat[["Life satisfaction"]]
```

```python
countries=life_sat["Country"]
```

# Plotting the GDP vs satisfaction index

```python
plt.rcParams["font.size"]=10
fig1,axis1=plt.subplots(figsize=(5,4))
axis1.set(xlabel="GDP per capita (USD)",ylabel="Life satisfaction",title="GDP vs life satisfaction factor")
axis1.scatter(GDP,life_sat_factor,label="life_sat")
```

```python
from sklearn.linear_model import LinearRegression 
```

# Select the Linear Regression model

```python
linear_model=LinearRegression()
```

# Train the model using LinearRegression()

```python
linear_model.fit(GDP,life_sat_factor)
```

# Predict using cyprus life satisafaction using LinearRegression()

```python
cyprus_gdp=[[37655.2],[37455.2]]
linear_model.predict(cyprus_gdp)
```

- predicting the life satisfaction of cyprus according using LinearRegression()

# Plotting the LinearRegression model

```python
fig,axis=plt.subplots()
axis.set(xlabel="GDP",ylabel="life satisfaction",title="prediction model of life satisfaction with GDP")
axis.scatter(GDP,life_sat_factor,c="blue",s=50)
axis.plot(GDP,linear_model.predict(GDP),label="Linear Regression model",
          c="grey",linewidth=5)
axis.scatter(cyprus_gdp,linear_model.predict(cyprus_gdp),
             marker="^",s=200,c="orange",label="cyprus life satisfaction prediction")
plt.legend()
plt.grid()
plt.show()
```

    - predicting the life satisfaction of cyprus according using KNeighborsRegressor
    

```python
from sklearn.neighbors import KNeighborsRegressor
```

# Select the KNeighbourRegressor model

```python
knn_model=KNeighborsRegressor(n_neighbors=5)
```

# Train the model using KNeighbourRegressor

```python
knn_model.fit(GDP,life_sat_factor)
```

# Prediction using KNeighboursRegressor model

```python
knn_model.predict(cyprus_gdp)
```

# Plotting the KNeighboursRegression model

```python
fig,axis=plt.subplots()
axis.scatter(GDP,life_sat_factor)
knn_model_prediction=knn_model.predict(GDP)
axis.set(xlabel="Gross Domestic Product",ylabel="Life satisfaction index",title="GDP vs Life satisfaction index")

axis.plot(GDP,knn_model_prediction,
          c="grey",linewidth=5,label="KNeighborRegression model")

plt.legend()
plt.grid()
plt.show()
```

# Comparing LinearRegression and KNeighorRegressor models

```python
fig,axis=plt.subplots(figsize=(10,9))
axis.set(xlabel="GDP",ylabel="Life satisfaction",title="Prediction model of life satisfaction with GDP")
axis.scatter(GDP,life_sat_factor,c="blue",s=50)
axis.plot(GDP,linear_model.predict(GDP),label="Linear Regression model",
          c="orange",linewidth=1)
axis.scatter(cyprus_gdp,linear_model.predict(cyprus_gdp),
             marker="^",s=500,c="green",label="cyprus life satifaction using LinearRegression")
axis.plot(GDP,knn_model_prediction,
          c="grey",linewidth=1,label="KNeighborRegression model")
axis.scatter(cyprus_gdp,knn_model.predict(cyprus_gdp),
             marker="x",s=500,c="magenta",label="cyprus life satifaction using k neighbor regression")
plt.legend()
plt.grid()
plt.show()
```
